```python
import numpy as np
%run "elegant_plot.ipynb" 

theta = np.linspace(-np.pi/3, np.pi/3,100) 
```

```python
import itertools
def flip(items, ncol):
    return itertools.chain(*[items[i::ncol] for i in range(ncol)])
```

```python
fig_save_path = r"\\storage01.amolf.nl\data\AMOLF\users\singh\Martin\Evolutionary Design\Thesis Chapter"\
                r"\Chapter 2.0\Figures\Non Flat Curves\Section 1"
```

## Family 1   - Linear $D_{\theta}$ Curves

```python
halffig()
ax = plt.gca()
ax.set_xticks([-np.pi/3, -np.pi/6, 0, np.pi/6, np.pi/3])
ax.set_xticklabels(['-60', '-30', '0', '30', '60'])
ax.set_xlim([-np.pi/2.5,np.pi/2.5])
plt.plot(theta, np.ones((100,1)), '--', color = 'black', linewidth = 0.5)

D = 0.25*(theta)/np.pi + 1
plt.plot(theta, D, color = '#CD0000', alpha = 1.0, label = r'$F1_1$')


D = 0.50*(theta)/np.pi + 1
plt.plot(theta, D, color = '#4c7d4c', alpha = 1.0, label = r'$F1_2$')

D = 0.75*(theta)/np.pi + 1
plt.plot(theta, D, color = '#2E348E', alpha = 1.0,  label = r'$F1_3$')

plt.legend( prop={'size':4.5}, frameon = False)

D = -0.25*(theta)/np.pi + 1
plt.plot(theta, D, color = '#CD0000', alpha = 0.4, label = r'$F1_4$')

D = -0.50*(theta)/np.pi + 1
plt.plot(theta, D, color = '#4c7d4c', alpha = 0.4, label = r'$F1_5$')

D = -0.75*(theta)/np.pi + 1
plt.plot(theta, D, color = '#2E348E', alpha = 0.4, label = r'$F1_6$')

handles, labels = ax.get_legend_handles_labels()
plt.legend( flip(handles, 3), flip(labels, 3),loc = 2, prop={'size':4.5}, frameon = False, bbox_to_anchor = (.15,0.99), ncol = 3)

plt.xlabel(r'$\theta^{\circ}$'); plt.ylabel('$D$', rotation = 0)
plt.ylim(0.7, 1.35)
#plt.suptitle(r'Family 1 - Linear $D_{\theta}$ Curves', size = 8, x = 0.59)
plt.savefig(fig_save_path + r"\Family1.pdf", format='pdf',bbox_inches='tight')
```

## Family 2   - Single Crest(Trough) Trigonometric $D_{\theta}$ Curves
